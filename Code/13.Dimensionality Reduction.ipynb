### Dimensionality Reduction

#### Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.manifold import TSNE
```

#### Data

```python
df = pd.read_csv('../Data/Digits.csv')
df.head()
```

```python
X = df.drop('y', axis=1).values
y = df['y'].values
```

**Split** the Data Set into Train Set and Test Set

```python
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size=0.2, 
                                                    random_state=0)
```

### Principal Component Analysis

```python
pca = PCA()
pca.fit(X_train, y_train)
X_transform = pca.transform(X_train)
plt.plot(pca.explained_variance_ratio_);
```

```python
sns.scatterplot(x = X_transform[:, 0], 
                y = X_transform[:, 1], 
                style = y_train, 
                hue = y_train, 
                palette = ['r','g','b']);
```

```python
print(f'Train Accuracy : {pca.score(X_train, y_train):.2f}')
```

```python
X_transform = pca.transform(X_test)
sns.scatterplot(x = X_transform[:, 0], 
                y = X_transform[:, 1], 
                style = y_test, 
                hue = y_test, 
                palette=['r','g','b']);
```

```python
print(f'Test Accuracy : {pca.score(X_test, y_test):.2f}')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size=0.2, 
                                                    random_state=0)
```

### Linear Discriminant Analysis 

```python
lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)
X_transform = lda.transform(X_train)
```

```python
sns.scatterplot(x = X_transform[:, 0], 
                y = X_transform[:, 1], 
                style = y_train, 
                hue = y_train, 
                palette = ['r','g','b']);
```

```python
print(f'Train Accuracy : {lda.score(X_train, y_train)*100:.2f}%')
```

```python
X_transform = lda.transform(X_test)
sns.scatterplot(x = X_transform[:, 0], 
                y = X_transform[:, 1], 
                style = y_test, 
                hue = y_test, 
                palette=['r','g','b']);
```

```python
print(f'Test Accuracy : {lda.score(X_test, y_test)*100:.2f}%')
```

### TSNE : t Distributed Stochastic Neighbor Embedding

### Perplexity = 1

How it Combines the High Dimensionality Data down to 2 Dimensionality

```python
tsne = TSNE(n_components=2, 
            perplexity=1, 
            random_state=0)

X_transform = tsne.fit_transform(X)

sns.scatterplot(x = X_transform[:,0], 
                y = X_transform[:, 1], 
                style = y, 
                hue = y, 
                palette = ['r','g','b']);
```
