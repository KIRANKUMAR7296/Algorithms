### Visualizing Decision Tree

### Library

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

### Dataset

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target
df.head()
```

### Splitting the Data Set into Train Set and Test Set

```python
X_train, X_test, y_train, y_test = train_test_split(df[iris.feature_names], 
                                                    df['species'], 
                                                    random_state=0)
```

### Decision Tree

```python
clf = DecisionTreeClassifier(max_depth=2, random_state=0)
clf.fit(X_train, y_train)
```

### Prediction on One Observation

```python
print(f'Prediction : {clf.predict(X_test.iloc[0].values.reshape(1,-1))}')
```

```python
print(f'Prediction on Top 10 : {clf.predict(X_test[:10])}')
```

### Performance of the Model

```python
score = clf.score(X_test, y_test)
print(f'Test Score : {score*100:.2f}%')
```

### Visualize Decision Tree using Matplotlib

Default Visualization

```python
 tree.plot_tree(clf); 
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(4,4), dpi=300)
tree.plot_tree(clf);
```

### More Interpretable Tree

```python
features = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width'] 
classes = ['Setosa', 'Versicolor', 'Virginica']

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(4,4), dpi=300)
tree.plot_tree(clf, 
               feature_names=features, 
               class_names=classes, 
               filled=True)
fig.savefig('../Image/Decision Tree.png');
```
